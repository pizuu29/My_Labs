### Build Command-Line Interfaces With Python's argparse

#### Command-Line Interfaces (CLIs)
Command-Line Interface (CLI) is a way to use and control a program by typing text commands instead of clicking buttons or using menus.

You type commands in a place like Command Prompt (Windows), Terminal (Mac/Linux), or PowerShell.

The program reads those commands and gives you a result in text form.

This is different from a Graphical User Interface (GUI), where you interact using your mouse and icons.

```python
import argparse
import sys

class Calculator:
    def add(self, x, y):
        return x + y
    
    def subtract(self, x, y):
        return x - y
    
    def multiply(self, x, y):
        return x * y
    
    def divide(self, x, y):
        if y == 0:
            return "Error: Division by zero!"
        return x / y

def main(custom_args=None):
    parser = argparse.ArgumentParser(description="Simple Calculator using argparse")
    parser.add_argument("operation", type=str, choices=["add", "sub", "mul", "div"], help="Operation to perform")
    parser.add_argument("num1", type=float, help="First number")
    parser.add_argument("num2", type=float, help="Second number")


        # Normal terminal execution
    args = parser.parse_args()

    calc = Calculator()

    if args.operation == "add":
        result = calc.add(args.num1, args.num2)
    elif args.operation == "sub":
        result = calc.subtract(args.num1, args.num2)
    elif args.operation == "mul":
        result = calc.multiply(args.num1, args.num2)
    elif args.operation == "div":
        result = calc.divide(args.num1, args.num2)

    print(f"Result: {result}")

if __name__ == "__main__":
    main()
```

The argparse module’s support for command-line interfaces is built around an instance of argparse.ArgumentParser. It is a container for argument specifications and has options that apply to the parser as whole:

parser = argparse.ArgumentParser(
                    prog='ProgramName',
                    description='What the program does',
                    epilog='Text at the bottom of help')

The ArgumentParser.add_argument() method attaches individual argument specifications to the parser. It supports positional arguments, options that accept values, and on/off flags:

```python
parser.add_argument('filename')           # positional argument
parser.add_argument('-c', '--count')      # option that takes a value
parser.add_argument('-v', '--verbose',
                    action='store_true')  # on/off flag
```

The ArgumentParser.parse_args() method runs the parser and places the extracted data in a argparse.Namespace object:

```python
args = parser.parse_args()
print(args.filename, args.count, args.verbose)
```
