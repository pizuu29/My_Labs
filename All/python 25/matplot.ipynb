```python
import csv
import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```

```python
data=pd.read_csv('data.csv')
```

Reads your CSV file named data.csv into a pandas DataFrame called data.

A DataFrame is like a spreadsheet in Python: it has rows and columns.

```python
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
```

ids: This is a column from your DataFrame containing each responders ID.

lang_responses: This is the column containing the programming languages each person works with.

Usually, lang_responses looks like strings such as "Python;Java;C++".

```python
language_counter = Counter()
```

Counter will help count how many times each language appears.

Initially, it is empty.

```python
for response in lang_responses:
    language_counter.update(response.split(';'))
```

Loop over each response in lang_responses. Example: "Python;Java;C++".

response.split(';')  splits the string into a list: ['Python', 'Java', 'C++'].

language_counter.update(list)  adds each language to the counter, counting how many times it occurs in total.

```python
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

language_counter.most_common(15)  gives the 15 most common languages as a list of (language, count) tuples. Example: [('Python', 200), ('Java', 150), ...].

Then we separate this into two lists:

languages  names of the languages

popularity  number of people using each language.

```python
languages.reverse()
popularity.reverse()
```

Reverses the lists because horizontal bar charts (barh) start plotting from the bottom.

This way, the most popular language appears at the top.

```python
plt.barh(languages, popularity)
plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")
plt.tight_layout()
plt.show()
```

barh  horizontal bar chart

languages  y-axis (vertical) labels

popularity  x-axis values (length of bars)

Sets the title of the chart.

Commented ylabel  y-axis is automatically the language names.

x-axis shows number of people using the language.

Makes sure that labels, title, and bars dont overlap.

# Stack Plot

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2', 'player3']
colors = ["#768a65", '#fc4f30', '#008fd5']

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)

plt.legend(loc=(0.07, 0.05))# this loc for the location of the legend by adjustment of 0.07 and 0.05 OR
#plt.legend(loc='lower left')
plt.title("My Awesome Stack Plot")
plt.tight_layout()
plt.show()
```

# Filling Area on Line Plots

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('data_1.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages,py_salaries,overall_median,alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

# Plotting Time Series Data

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn-v0_8')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
plt.plot_date(dates,y,linestyle='solid')# The line style give the line on marker
plt.gcf().autofmt_xdate()# gcf mean get current formate and autofmt_xdate give the date as rotate on the graph.
date_formate=mpl_dates.DateFormatter('%b, %d %Y')# this will give the format of date %b month %d day %y year 
plt.gca().xaxis.set_major_formatter(date_formate)# gca mean get current axis the current axis give on xaxis with date_format
plt.tight_layout()

plt.show()
```

```python
#  This code well not run in jupyter note book run this code in .py file
import random
from itertools import count
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []
index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    plt.cla()
    plt.plot(x_vals, y_vals, label="Random Data")
    plt.legend(loc="upper left")
    plt.tight_layout()

fig, ax = plt.subplots()
ani = FuncAnimation(fig, animate, interval=1000, cache_frame_data=False)

HTML(ani.to_jshtml())   # <- shows animation inside notebook
```
