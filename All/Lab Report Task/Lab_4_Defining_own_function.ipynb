In circuit design, you often calculate series resistance.
1.	Write a function series() that takes a list or tuple of resistances.
2.	Return the total resistance.
3.	Add docstrings and annotations.
4.	Print the function‚Äôs `__annotations__.`


```python
# def series(resistances: list[float] | tuple[float, ...]) -> float:
def series(resistances):
    """
    Calculate the total resistance of resistors connected in series.

    Parameters:
        resistances (list[float] or tuple[float, ...]):
            A collection of resistor values in ohms.

    Returns:
        float: The total resistance in ohms.
    """
    return sum(resistances)

# Print annotations
print(series.__annotations__)

# Example usage
print("series([10, 20, 30]) ->", series([10, 20, 30]), "ohms")
print("series((1.5, 2.5, 3.0)) ->", series((1.5, 2.5, 3.0)), "ohms")
print("series([]) ->", series([]), "ohms")
 
```
